# Filtering Data

```python
# Output only the rows with type 1 == Grass
poke_df.loc[poke_df['Type 1'] == 'Grass']
```

```python
# Output rows with Type 1 grass and type 2 poison
poke_df.loc[(poke_df['Type 1'] == 'Grass') &
            (poke_df['Type 2'] == 'Poison') &
           (poke_df['HP'] > 70)]
# Make sure the 2 conditions are separated by parentheses
# Also the conditional operators inside pandas parameters should be:
# & for and; | for or
```

```python
# Copy these values to another dataframe:
grass_poison_hp70 = poke_df.loc[(poke_df['Type 1'] == 'Grass') &
            (poke_df['Type 2'] == 'Poison') &
           (poke_df['HP'] > 70)]

grass_poison_hp70
```

However, the index of the new dataframe are preserved from its parent dataframe.

```python
# To reset index
grass_poison_hp70.reset_index(drop=True, inplace=True)
grass_poison_hp70
```

```python
# You can export the new dataframe into a csv file as a checkpoint
grass_poison_hp70.to_csv('grass-poison-hp70.csv', index=False)
```

Returning to the poke_df dataframe:

```python
# If you want to retain only the 'mega' pokemons:
poke_df.loc[poke_df['Name'].str.contains('Mega')]
```

```python
# If you want to filter out 'Mega' name:
poke_df.loc[~poke_df['Name'].str.contains('Mega')]
# ~ (tilde) is a NOT conditional operator
```

```python
# If you want to get the fire and grass type pokemons only
# You can use regex:
import re

poke_df.loc[poke_df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)]
# flags=re.I means ignore capitalization
# regex=True means you enable regex symbols/functions inside the contains function
```

```python
# If you want to find out all the pokemon names that contain the
# string sequence: "pi"
poke_df.loc[poke_df['Name'].str.contains('pi[a-z]*', flags=re.I, regex=True)]
# [a-z] range of letters from a to z
# asterisk (*) at the end means whatever sequence of [a-z] from zero to
# infinite amount of characters
```

```python
# If you want to find every pokemon that has a name that starts with
# 'pi'
poke_df.loc[poke_df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)]
```

## Aggregate Statistics (groupby function)
poke_df is a dataframe. You can group it accordingly using groupby('column_name') function. If you want to group it according to multiple columns, use DataFrame.groupby(['column_name1','column_name2'])

Grouping by Type 1 and finding the mean of each type 1 then sorting their values by Defense in descending order to find out which type of pokemon has the highest amount of Defense

```python
poke_df.groupby(['Type 1']).mean().sort_values('Defense', ascending=False)
```

Establishing a count column and grouping by Type 1 and finding the total number of pokemons inside each Type 1

```python
poke_df['count'] = 1
poke_df.groupby(['Type 1']).count()['count']
# It does not count rows with NaN (null type values)
```

If you want to group it by multiple columns:

```python
poke_df.groupby(['Type 1', 'Type 2']).count()['count']
```
