```python

def assign_session(df):
    # first session starts with 1
    session_number = 1
    
    # create an empty list to store the session numbers
    session_list = []
    
    # go through each row
    for ind, row in df.iterrows():
        session_list.append(session_number)
        # if the value of conversation status is closed - 
        if row['conversation_status'] == 'closed':
            session_number = session_number + 1
            
    # session_list will have a value of [1,1,1,1,2,3]
    # e.g. where the length of the list = length of df
    df['session'] = session_list
            
    return df 


```

```python

# create an empty list
answer_list = []

# loop to apply across conversation

for ind, df in chat_df.groupby('conversation_id'):
    # run the function return a dataframe
    df = assign_session(df) 
    
    # list will have a list of dataframes
    answer_list.append(df)
    
# we will have an answer_list with a list of dataframe
    
# combine the list of dataframes into one dataframe
answer_df = pd.concat(answer_list)
```
