# Using PivotTable

### How many messagees are attributed to each Shopee shop?

```python
chat_pivot = chat_df.pivot_table(index='shopid', values='message_id', aggfunc='count')
chat_pivot.sort_values('message_id', ascending=False).reset_index()
```

### How many messages each buyer sent in or receive in total?

```python
top_buyers = chat_df.pivot_table(index='buyer_userid', values='message_id', aggfunc='count')
top_buyers = top_buyers.sort_values('message_id', ascending=False).reset_index()
```

### Remove from the dataset, messages from the top 3 buyers by messages?

```python
top_3_buyers = top_buyers.head(3)
top_3_buyers
```

## Workshop Question (1)

```python
# Use convers.csv for a smaller table
chat_df = pd.read_csv('Chats.csv')
test_df = chat_df[chat_df['conversation_id'] == 'C249977464015359203']
test_df
```

```python
def assign_session(df):
    session_number = 1
    
    # create an empty list to store the session numbers
    session_list = []
    for ind, row in df.iterrows():
        session_list.append(session_number)
        # if the value of conversation status is closed.
        if row['conversation_status'] == 'closed':
            session_number += 1
    df['session'] = session_list
    
    return df
```

```python
chat_df.groupby('conversation_id')
# create an answer_list
answer_list = []
for ind, df in chat_df.groupby('conversation_id'):
    # run the function return a dataframe
    df = assign_session(df)
    
    # list will have a list of dataframes
    answer_list.append(df)

answer_df = pd.concat(answer_list)
```

```python
answer_df
```

## Workshop Question (2)

How long will the seller reply to the first message of the buyer?

```python
all_df = []

chat_df['timestamp'] = pd.to_datetime(chat_df['timestamp'])

# get session id by combining conversation id and message id
chat_df['session_id'] = chat_df['conversation_id'].astype(str) + "_" + answer_df['session'].astype(str)

for ind, df in chat_df.groupby('session_id'):
    df = df.sort_values('timestamp')
    # get a dataframe of only buyers
    buyer_df = df[df['is_buyer']]
    if len(buyer_df) == 0:
        # outbound
        df['first_reply_time'] = -2
    else:
        # get earliest message from buyer
        earliest_buyer_message = buyer_df['timestamp'].min()
        
        # filter out seller messages that are before the earliest buyer message
        
        seller_timing = df[df['is_seller']]
        seller_timing = seller_timing[seller_timing['timestamp'] > earliest_buyer_message]
        
        if len(seller_timing) == 0:
            #no reply backlog
            df['first_reply_time'] = -1
        
        else:
            seller_earliest = seller_timing['timestamp'].min()
            time_diff = seller_earliest - earliest_buyer_message
            df['first_reply_time'] = time_diff.seconds
    all_df.append(df)

chat_df = pd.concat(all_df)
```
