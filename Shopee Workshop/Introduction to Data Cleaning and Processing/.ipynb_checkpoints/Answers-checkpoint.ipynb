```python
csat_df = pd.read_csv('csat.csv')
csat_df.head()
```

```python
# on = 'csat_id' means match on the csat_id column
# how = 'left' - keep all rows from the left table(df)

# you will encounter this error - 
# ValueError: You are trying to merge on object and int64 columns. If you wish to proceed you should use pd.concat
# Thats because the csat_df's data is a integer type, while df's csat data is a string
# convert csat_df data into a string

csat_df['csat_id'] = csat_df['csat_id'].astype(str)
chat_df = chat_df.merge(csat_df, on = 'csat_id', how = 'left')

# df
chat_df[~chat_df['grade'].isnull()].loc[:,['message_id','content','grade']].head()
```

```python





# Adding sessions

# Create a function

def assign_session(df):
    # first session starts with 1
    session_number = 1
    
    # create an empty list to store the session numbers
    session_list = []
    
    # go through each row
    for ind, row in df.iterrows():
        session_list.append(session_number)
        if row['conversation_status'] == 'closed':
            session_number = session_number + 1
            
    # session_list will have a value of [1,1,1,1,2,3] e.g.
    df['session'] = session_list
            
    return df 


```

```python
# apply it accross each conversation_id

# create an empty list
df_all = []

# go through each segmented dataframe through conversation id
# note, remember not to use df again, use df_c or other variables

for ind, df_c in chat_df.groupby('conversation_id'):
    
    # call the function you just created
    df_c = assign_session(df_c)
    
    # append the processed segmented dataframe into assign_session list
    df_all.append(df_c)

# use pd.concat to combine the segmented dataframes back
chat_df = pd.concat(df_all)
```

# 1) Get the product id for each conversation from the product_url column
- Product id is the numbers after â€œpid=â€
- assign to column pid



```python
# e.g. "'<img class="gb_4b" src="www.shopeeprod.com/pid=P76540photo.jpg" alt="" aria-hidden="true" data-atf="0">'" 
# convert to to 76540

# Create an empty list to store values 
photo_imgs = []

# Repeat this step for each value in content
# ind is the index(row number), row is the row itself
for ind, row in chat_df.iterrows():
    # To access the 'content' value  - row['content']
    
    # Check if inside the content value, there is the string 'rate_cards'
    if 'pid' in row['photo_img']:
        # apply the logic we used above
        photo = row['photo_img'].split('pid=')[1].split('photo')[0]
    else:
        # if it doesnt have 'rate_cards in the content assign a blank string'
        photo = ''
        
    # push the value into the empty list
    photo_imgs.append(photo)
    

# assign this list as a new column csat_id
chat_df['pid'] = photo_imgs    
    
# Check those rows with csat_data
chat_df[chat_df['pid'] != ''].loc[:,['message_id','photo_img','pid']].head()
```

# 2) Flag all conversations that doesn't have a seller (is_seller = True) with the column missed_chat = True
# Remove all these data



```python
# get session id by combining conversation id and sessionid
chat_df['session_id'] = chat_df['conversation_id'].astype(str) + "_" + chat_df['session'].astype(str)
```

```python
all_df = []


for ind, df in chat_df.groupby('session_id'):
    # check if any of the messages are from sellers
    if any(df['is_seller']):
        df['missed'] = False
        all_df.append(df)
    else:
        df['missed'] = True
        
    
        
chat_df = pd.concat(all_df)
chat_df = chat_df[~chat_df['missed']]
```

# Assign a new column -> Chat Duration, the time difference between the first chat and the last chat for each session.

- 	One session will all have the same value of â€˜chat_durationâ€™


```python
all_df = []
chat_df['timestamp'] = pd.to_datetime(chat_df['timestamp'])

for ind, df in chat_df.groupby('session_id'):
    # find the latest message
    max_time = df['timestamp'].max()
    
    # find the earliest message
    min_time = df['timestamp'].min()
    
    # get the time difference
    time_diff = max_time - min_time
    
    # Find the total chat_duration by taking the difference and add it to the column
    df['chat_duration'] = time_diff.seconds
    
    # append it to the list
    all_df.append(df)

# all_df will now have a list of dataframes
chat_df = pd.concat(all_df)
```

# Assign a new column -> first reply time in seconds.

## The reply time for each chat is the time taken from the first buyer message and the next seller message after the first buyer message.

## One session will all have the same value of â€˜first_reply_timeâ€™


```python
all_df = []

chat_df['timestamp'] = pd.to_datetime(chat_df['timestamp'])

# get session id by combining conversation id and sessionid
chat_df['session_id'] = chat_df['conversation_id'].astype(str) + "_" + chat_df['session'].astype(str)

for ind, df in chat_df.groupby('session_id'):
    df = df.sort_values('timestamp')
    # get a dataframe of only buyers
    buyer_df = df[df['is_buyer']]
    
    if len(buyer_df) == 0:
        # outbound
        df['first_reply_time'] = -2
    else:
        # get earliest message from buyer
        earliest_buyer_message = buyer_df['timestamp'].min()

        # Filter out seller messages that are before the earliest buyer message
        
        seller_timing = df[df['is_seller']]
        seller_timing = seller_timing[seller_timing['timestamp']  > earliest_buyer_message]
    
        if len(seller_timing) == 0:
            # No reply Backlog
            df['first_reply_time'] = -1
        else:
            
            seller_earliest = seller_timing['timestamp'].min()
            time_diff = seller_earliest - earliest_buyer_message
            df['first_reply_time'] = time_diff.seconds
    all_df.append(df)

            
chat_df = pd.concat(all_df)
```
