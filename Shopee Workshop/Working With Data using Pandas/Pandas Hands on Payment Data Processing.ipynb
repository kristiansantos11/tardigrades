### Is cancellation rate higher for certain payment channels?

```python
orders = pd.read_excel('sample_orders.xlsx', sheet_name = 'Order Data')
payments = pd.read_excel('sample_orders.xlsx', sheet_name = 'Payments Data')
```

```python
orders.drop(orders[orders['shopid'] == 111111111].index, axis=0, inplace=True)
```

```python
orders
```

```python
payments
```

```python
orders_with_pmt = orders.merge(payments[['checkoutid','payment_method']], on='checkoutid', how='left')
orders_with_pmt
```

```python
orders_with_pmt['cancel_status'] = 'Not Cancelled'
orders_with_pmt.loc[orders_with_pmt['status'].isin([6,7,8,9,10]), 'cancel_status'] = 'Cancelled'
orders_with_pmt
```

```python
cancel_by_pmt = orders_with_pmt.groupby(['payment_method', 'cancel_status'])['orderid'].nunique()
round((cancel_by_pmt/cancel_by_pmt.sum(level="payment_method"))*100, 2)
```

### How many percent of orders are re-ordered from the same shop, and how long after cancellation do they re-order (mean)?

```python
cancelled_orders = orders[orders['status'].isin([6,7,8,9,10])]
not_cancelled_orders = orders[~orders['status'].isin([6,7,8,9,10])]
```

```python
cancelled_orders
```

```python
not_cancelled_orders
```

```python
compare = cancelled_orders.merge(not_cancelled_orders, how='inner', on=['userid', 'shopid'])

# _x is for cancelled_orders dataset and _y is for not_cancelled_orders dataset
compare
```

```python
compare['created_time_y'] = pd.to_datetime(compare['created_time_y'])
reorder = compare[compare['created_time_y'] >= compare['cancel_time_x']]
reorder
```

```python
reorder_time = (reorder['created_time_y'] - reorder['cancel_time_x']).mean()
```

```python
print(f"Total number of orders re-ordered after cancellation: {str(reorder['orderid_x'].nunique())}")
print(f"Mean of how long after cancellation do they reorder: {str(reorder_time)}")
```

```python
# Optional: If you look at the dataset, there are actually some orders re-ordered after 1 week, which are anomalies in this case
# We can limit the time difference and re-calculate
import datetime as dt

reorder_2 = reorder[(reorder['created_time_y'] - reorder['cancel_time_x']).dt.days < 7]
reorder_time_2 = (reorder_2['created_time_y'] - reorder_2['cancel_time_x']).mean()
```

```python
print("Total number of orders re-ordered after cancellation: " + str(reorder_2['orderid_x'].nunique()))
print("Average time buyers take to re-order after cancellation: " + str(reorder_time_2))
```
